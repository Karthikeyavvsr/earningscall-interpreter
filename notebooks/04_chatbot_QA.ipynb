```python
from transformers import pipeline

# Load the Question Answering pipeline
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
```

```python
# Example: hardcoded or loaded from a file
transcript_text = """
Good morning, everyone. This is the Roots Q4 and FY2024 earnings call, hosted by CEO Megan Roach and CFO Leon Wu. The company reported strong momentum in Q4, with total sales reaching $110.8 million, up 2.4% year-over-year (YoY), and up 4.5% YoY excluding the extra week in 2023. This growth was primarily driven by the Direct-to-Consumer (DTC) segment, where comparable sales increased by 7.5%, marking the highest comp sales growth since 2017.

Gross margin expanded by 270 basis points to 61.3%, reflecting improved product cost optimization and reduced discounting. Adjusted EBITDA rose to $25.3 million, or 22.8% of sales, up from $23.2 million last year. Net debt was significantly reduced to $7.3 million, a 56.7% decline from 2023 levels and down from a peak of over $96 million in 2019.

Key drivers of this performance included AI-driven inventory allocation, optimized marketing investments, and strong engagement through the “Anything Roots, Everything Quality” campaign. Experiential activations and influencer partnerships contributed to brand strength and consumer reach. Products such as the Cooper Fleece, Cloud Fleece, and the new Activewear collection performed strongly, with activewear sales up 40% YoY.

From an operational standpoint, Roots completed the first phase of implementing AI tools for inventory and online personalization. These tools improved product discovery, tailored interactions, and supported higher conversion rates. The company also leveraged a newly established data warehouse to drive better customer insights and demand forecasting.

Leon Wu discussed financial highlights, noting DTC sales of $101.2 million, up 6% YoY when excluding the extra week. Gross profit was $68 million, up from $63.4 million, while total gross margin for the year was 61.3%. Selling, general, and administrative expenses were $45.2 million, with increases attributed to the absence of one-time gains and higher marketing spend.

A non-cash impairment on intangible assets resulted in an accounting net loss of $21.7 million for Q4. Excluding the impairment, net income would have been $15 million or $0.37 per share, up 2.9% from last year. Full-year adjusted EBITDA was $21.3 million, up from $19.9 million in 2023.

Inventory at year-end was $41 million, up 13.4% YoY, driven by restocking core styles and supporting spring 2025. Free cash flow for the quarter was $39.4 million, up 9.3% YoY.

Looking ahead, Roots plans to continue investments in marketing, brand ambassadors, AI, and retail experience upgrades. The company also announced a normal course issuer bid (NCIB) to repurchase up to 1.3 million shares, reflecting confidence in long-term value creation.

Megan Roach concluded by reaffirming Roots’ strong start to Q1 2025 with low double-digit DTC comp sales growth and confidence in the brand’s strategy, despite economic uncertainty.


"""

# Keep it short for testing (~500 words); full transcripts may need chunking
```

```python
# Ask a natural language question
question = "What was said about revenue?"

# Run QA
result = qa_pipeline({
    'question': question,
    'context': transcript_text
})

print(f"🤖 Answer: {result['answer']}")
```

```python
questions = [
    "What was said about revenue?",
    "Did they mention profit margins?",
    "Any comments on product launches?",
    "What guidance was provided for next quarter?"
]

for q in questions:
    result = qa_pipeline({'question': q, 'context': transcript_text})
    print(f"Q: {q}\nA: {result['answer']}\n")
```
